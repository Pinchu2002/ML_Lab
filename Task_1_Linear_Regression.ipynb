```python
X = np.array(advertising['TV']).reshape(-1,1)
y = np.array(advertising['Sales']).reshape(-1,1)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, test_size=0.25, random_state=100)
Linear_Reg = LinearRegression()
Linear_Reg.fit(X_train, y_train)
print("Linear_Reg_Score : ", Linear_Reg.score(X_test, y_test))
y_pred = Linear_Reg.predict(X_test)
plt.scatter(X_test, y_test, color='b')
plt.plot(X_test, y_pred, color='k')
plt.show()
```

```python
#Building the Model
def gradient_descent(X, y, learning_rate=0.000001, iterations=1000):
    # Initialize parameters
    m_curr = b_curr = 0
    n = len(X)
    
    for i in range(iterations):
        # Calculate predicted values
        y_predicted = m_curr * X + b_curr
        
        # Calculate gradients (derivatives) of the cost function with respect to m and b
        m_derivative = (-2/n) * np.sum(X * (y - y_predicted))
        b_derivative = (-2/n) * np.sum(y - y_predicted)
        
        # Update m and b using the gradients and learning rate
        m_curr = m_curr - learning_rate * m_derivative
        b_curr = b_curr - learning_rate * b_derivative
        
        # Calculate cost
        cost = (1/2) * np.sum((y - y_predicted) ** 2)
        
        # Print progress
        if i % 10 == 0:
            print("Iteration {}, Cost: {:.4f}".format(i, cost))
    return m_curr, b_curr

m, b = gradient_descent(X_train, y_train)
print("Optimized slope (m):", m)
print("Optimized intercept(b):",b)
```

```python
# Scatter plot of the training data
plt.scatter(X_train, y_train, color='blue', label='Training Data')

# Plot the linear regression line
y_pred = m * X_train + b
plt.plot(X_train, y_pred, color='red', label='Linear Regression')

# Labeling the plot
plt.xlabel('Feature (X_train)')
plt.ylabel('Target (y_train)')
plt.title('Linear Regression with Gradient Descent')
plt.legend()

# Display the plot
plt.show()
```
